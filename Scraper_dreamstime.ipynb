```python
import unittest, time, random
import urllib.request
from selenium import webdriver
from webdriver_manager.firefox import GeckoDriverManager
from bs4 import BeautifulSoup
import pandas as pd
```

```python
links = []
imgs = []
website = "https://www.stockvault.net//"
```

```python
rate = [i/10 for i in range(10)]
cnt = 0

quote = '"'
newline = '\n'
colon = ' : '
```

```python
browser = webdriver.Firefox(executable_path="C:\\Users\\meghc\\.wdm\\drivers\\geckodriver\\win64\\v0.29.1\\geckodriver.exe")
```

```python
browser.get("https://www.stockvault.net/free-photos/grape")
content = browser.page_source
soup1 = BeautifulSoup(content, "html.parser")
```

```python
soup1
```

```python
soup2 = soup1.find('div', attrs = {'id' : 'flexgrid'})
for main in soup2.findAll('div', attrs = {'class' : 'item'}):
    
    imgl = main.find('img', src=True)
    if (imgl != ''):
        imgs.append((imgl.get('src')).strip())
```

```python
print("Got imgs : ", len(imgs))
```

```python
imgs
```

```python
for img in imgs:
    img = website + img[4:]
#    print(img)
    urllib.request.urlretrieve(img, str(cnt+122) + '.jpg')
    time.sleep(random.choice(rate))
    cnt = cnt + 1
    if cnt == 1:
        break
```
